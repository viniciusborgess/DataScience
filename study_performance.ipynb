```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
text_file = pd.read_csv('study_performance.csv', encoding='utf-8')
```

```python
df = text_file
```

```python
df
```

```python
# 1. Visualização da distribuição dos Scores de matemática, leitura e escrita
plt.figure(figsize=(10, 6))

plt.subplot(1, 3, 1)
plt.hist(df['math_score'], color='skyblue', edgecolor='black')
plt.title('Distribuição dos Escores de Matemática')
plt.xlabel('Pontuação')
plt.ylabel('Frequência')

plt.subplot(1, 3, 2)
plt.hist(df['reading_score'], color='salmon', edgecolor='black')
plt.title('Distribuição dos Escores de Leitura')
plt.xlabel('Pontuação')
plt.ylabel('Frequência')

plt.subplot(1, 3, 3)
plt.hist(df['writing_score'], color='lightgreen', edgecolor='black')
plt.title('Distribuição dos Escores de Escrita')
plt.xlabel('Pontuação')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
# 2. Calculando as médias dos Scores de matemática, leitura e escrita
math_mean = df['math_score'].mean()
reading_mean = df['reading_score'].mean()
writing_mean = df['writing_score'].mean()

# Visualizando as médias
plt.figure(figsize=(8, 5))

labels = ['Matemática', 'Leitura', 'Escrita']
means = [math_mean, reading_mean, writing_mean]

plt.bar(labels, means, color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Média dos Escores por Disciplina')
plt.xlabel('Disciplina')
plt.ylabel('Média')
plt.ylim(0, 100)

plt.show()
```

```python
# 3. Distribuição dos Scores de matemática por nível educacional dos pais
plt.figure(figsize=(10, 6))

edu_levels = df['parental_level_of_education'].unique()

for edu_level in edu_levels:
    subset = df[df['parental_level_of_education'] == edu_level]
    plt.hist(subset['math_score'], bins=10, alpha=0.5, label=edu_level)

plt.title('Distribuição dos Escores de Matemática por Nível Educacional dos Pais')
plt.xlabel('Pontuação')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
# 4. Relação entre o tipo de almoço e o desempenho em matemática
plt.figure(figsize=(8, 5))

plt.bar(df['lunch'], df['math_score'], color=['lightblue', 'lightcoral'])
plt.title('Desempenho em Matemática por Tipo de Almoço')
plt.xlabel('Tipo de Almoço')
plt.ylabel('Pontuação em Matemática')

plt.show()
```

```python
# 5. Distribuição dos Scores de leitura por gênero
plt.figure(figsize=(8, 5))

female_scores = df[df['gender'] == 'female']['reading_score']
male_scores = df[df['gender'] == 'male']['reading_score']

plt.hist([female_scores, male_scores], color=['lightcoral', 'lightblue'], alpha=0.6, label=['Feminino', 'Masculino'])
plt.title('Distribuição dos Escores de Leitura por Gênero')
plt.xlabel('Pontuação em Leitura')
plt.ylabel('Frequência')
plt.legend()

plt.show()
```

```python
# 6. Desempenho em matemática por grupo étnico
plt.figure(figsize=(10, 6))

ethnic_groups = df['race_ethnicity'].unique()

for ethnic_group in ethnic_groups:
    subset = df[df['race_ethnicity'] == ethnic_group]
    plt.hist(subset['math_score'], bins=10, alpha=0.5, label=ethnic_group)

plt.title('Desempenho em Matemática por Grupo Étnico')
plt.xlabel('Pontuação')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
# 7. Distribuição conjunta dos Scores de escrita e leitura
plt.figure(figsize=(8, 6))

plt.scatter(df['writing_score'], df['reading_score'], color='lightgreen', alpha=0.7)
plt.title('Distribuição Conjunta dos Escores de Escrita e Leitura')
plt.xlabel('Pontuação em Escrita')
plt.ylabel('Pontuação em Leitura')

plt.show()
```
