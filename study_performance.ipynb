```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
text_file = pd.read_csv('study_performance.csv', encoding='utf-8')
```

```python
df = text_file
```

```python
df
```

```python
# 1. VisualizaÃ§Ã£o da distribuiÃ§Ã£o dos Scores de matemÃ¡tica, leitura e escrita
plt.figure(figsize=(10, 6))

plt.subplot(1, 3, 1)
plt.hist(df['math_score'], color='skyblue', edgecolor='black')
plt.title('DistribuiÃ§Ã£o dos Escores de MatemÃ¡tica')
plt.xlabel('PontuaÃ§Ã£o')
plt.ylabel('FrequÃªncia')

plt.subplot(1, 3, 2)
plt.hist(df['reading_score'], color='salmon', edgecolor='black')
plt.title('DistribuiÃ§Ã£o dos Escores de Leitura')
plt.xlabel('PontuaÃ§Ã£o')
plt.ylabel('FrequÃªncia')

plt.subplot(1, 3, 3)
plt.hist(df['writing_score'], color='lightgreen', edgecolor='black')
plt.title('DistribuiÃ§Ã£o dos Escores de Escrita')
plt.xlabel('PontuaÃ§Ã£o')
plt.ylabel('FrequÃªncia')

plt.tight_layout()
plt.show()
```

```python
# 2. Calculando as mÃ©dias dos Scores de matemÃ¡tica, leitura e escrita
math_mean = df['math_score'].mean()
reading_mean = df['reading_score'].mean()
writing_mean = df['writing_score'].mean()

# Visualizando as mÃ©dias
plt.figure(figsize=(8, 5))

labels = ['MatemÃ¡tica', 'Leitura', 'Escrita']
means = [math_mean, reading_mean, writing_mean]

plt.bar(labels, means, color=['skyblue', 'salmon', 'lightgreen'])
plt.title('MÃ©dia dos Escores por Disciplina')
plt.xlabel('Disciplina')
plt.ylabel('MÃ©dia')
plt.ylim(0, 100)

plt.show()
```

```python
# 3. DistribuiÃ§Ã£o dos Scores de matemÃ¡tica por nÃ­vel educacional dos pais
plt.figure(figsize=(10, 6))

edu_levels = df['parental_level_of_education'].unique()

for edu_level in edu_levels:
    subset = df[df['parental_level_of_education'] == edu_level]
    plt.hist(subset['math_score'], bins=10, alpha=0.5, label=edu_level)

plt.title('DistribuiÃ§Ã£o dos Escores de MatemÃ¡tica por NÃ­vel Educacional dos Pais')
plt.xlabel('PontuaÃ§Ã£o')
plt.ylabel('FrequÃªncia')
plt.legend()
plt.show()
```

```python
# 4. RelaÃ§Ã£o entre o tipo de almoÃ§o e o desempenho em matemÃ¡tica
plt.figure(figsize=(8, 5))

plt.bar(df['lunch'], df['math_score'], color=['lightblue', 'lightcoral'])
plt.title('Desempenho em MatemÃ¡tica por Tipo de AlmoÃ§o')
plt.xlabel('Tipo de AlmoÃ§o')
plt.ylabel('PontuaÃ§Ã£o em MatemÃ¡tica')

plt.show()
```

```python
# 5. DistribuiÃ§Ã£o dos Scores de leitura por gÃªnero
plt.figure(figsize=(8, 5))

female_scores = df[df['gender'] == 'female']['reading_score']
male_scores = df[df['gender'] == 'male']['reading_score']

plt.hist([female_scores, male_scores], color=['lightcoral', 'lightblue'], alpha=0.6, label=['Feminino', 'Masculino'])
plt.title('DistribuiÃ§Ã£o dos Escores de Leitura por GÃªnero')
plt.xlabel('PontuaÃ§Ã£o em Leitura')
plt.ylabel('FrequÃªncia')
plt.legend()

plt.show()
```

```python
# 6. Desempenho em matemÃ¡tica por grupo Ã©tnico
plt.figure(figsize=(10, 6))

ethnic_groups = df['race_ethnicity'].unique()

for ethnic_group in ethnic_groups:
    subset = df[df['race_ethnicity'] == ethnic_group]
    plt.hist(subset['math_score'], bins=10, alpha=0.5, label=ethnic_group)

plt.title('Desempenho em MatemÃ¡tica por Grupo Ã‰tnico')
plt.xlabel('PontuaÃ§Ã£o')
plt.ylabel('FrequÃªncia')
plt.legend()
plt.show()
```

```python
# 7. DistribuiÃ§Ã£o conjunta dos Scores de escrita e leitura
plt.figure(figsize=(8, 6))

plt.scatter(df['writing_score'], df['reading_score'], color='lightgreen', alpha=0.7)
plt.title('DistribuiÃ§Ã£o Conjunta dos Escores de Escrita e Leitura')
plt.xlabel('PontuaÃ§Ã£o em Escrita')
plt.ylabel('PontuaÃ§Ã£o em Leitura')

plt.show()
```
